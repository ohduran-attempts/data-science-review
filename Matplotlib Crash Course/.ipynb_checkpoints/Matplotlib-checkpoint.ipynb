# Matplotlib

matplotlib.org   /gallery

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x, y, 'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented method
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.2, 0.4, 0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('Big')
axes2.set_title('Small')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for i in range(len(axes)):
    axes[i].plot(x,y)
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2), dpi=100) #dotsperinch

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,7))

axes[0].plot(x,y)
axes[0].set_title('Y vs X')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y,x)
axes[1].set_title('X vs Y')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')

plt.tight_layout() #avoid overlapping
```

```python
#save a figure

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend()
```

```python
#fig.savefig('picture.png', dpi=200) #png, jpg, pdf are valid.
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#RGB hex code, line width, transparency, linestyle, mark for each point, their color, 
# border width and its color, 
ax.plot(x,y, color='#FF0033', linewidth=3, alpha=0.5, linestyle='--', marker='o', markersize=15,
       markerfacecolor='purple', markeredgewidth=3, markeredgecolor='blue')

ax.set_xlim([0, 1]) #min and max X
ax.set_ylim([0, 2]) #min and max Y
```
